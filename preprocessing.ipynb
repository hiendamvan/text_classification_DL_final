# Preprocessing data

```python
from pyvi import ViTokenizer, ViPosTagger
import numpy as np
import gensim
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/data.csv')
```

```python
df.head()
```

```python
# Remove punctuations
import string
def cleaner_update(text):
    return text.translate(str.maketrans('', '', string.punctuation))
```

```python
clean_test = []
for i in range(0,len(df)):
    clean_test.append(cleaner_update(df['Paragraph'][i]))
```

```python
# Tokenization using ViTokenizer
prep = []
for i in range(len(clean_test)):
    line = clean_test[i]
    lines = gensim.utils.simple_preprocess(line)
    lines = ' '.join(lines)
    lines = ViTokenizer.tokenize(lines)

    prep.append(lines)
```

```python
df['Topic'].unique()
```

```python
# Create a mapping from category to number
topic_mapping = {topic: i for i, topic in enumerate(df['Topic'].unique())}

# Map the 'Topic' column to numbers
df['Topic'] = df['Topic'].map(topic_mapping)

df['Topic'].unique()
```

```python
df['Paragraph'] = prep
```
